```python
import numpy as np
A=np.loadtxt("EtudeEmbeddings/gdn_common_sorted_Uberfiltered.txt", usecols=range(1,701))
B=np.loadtxt("EtudeEmbeddings/frwac_common_sorted_Uberfiltered.txt", usecols=range(1,701))
```

```python
#import torch

import scipy
import scipy.linalg as sl

M=np.matmul(np.transpose(B),A)

U, S, V_t = sl.svd(M, full_matrices=True)

W=np.matmul(U,V_t)
```

```python
result=list()
from scipy import spatial
for i in range(A.shape[0]):
    result.append(1 - spatial.distance.cosine(A[i], B[i]))
```

```python
x=np.loadtxt("EtudeEmbeddings/frwac_common_sorted_filtered.txt", usecols=range(1,701))
A=np.loadtxt("EtudeEmbeddings/gdn_common_sorted_filtered.txt", usecols=range(1,701))
B=np.matmul(x,W)

dico={}
result=list()
from scipy import spatial
for i in range(A.shape[0]):
    calcul = 1 - spatial.distance.cosine(A[i], B[i])
    result.append(calcul)
    if calcul < 0.1:
        dico[i]=calcul
```

```python
result
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.hist(result)
plt.savefig("procruste_similarity_histogram2.png")
```
